```python
import pandas as pd
import numpy as np

# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 50, 40],
    'Salary': [30000, 34000, 50000, 29000]
}
df = pd.DataFrame(data)
print(df)

# Reading from CSV (commented out here)
# df = pd.read_csv('data.csv')

# Display first 2 rows
print(df.head(2))

# Summary statistics
print(df.describe())

# Average Age
print(df['Age'].mean())

# Data Selection
print(df[df['Age'] > 30])  # Rows where age > 30
print(df.loc[1:2, ['Name', 'Salary']])  # Rows 1 to 2, columns 'Name' and 'Salary'

# Handling missing data
df_with_nan = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8]
})

# Replacing NaNs with 0
print(df_with_nan.fillna(value=0))

# Dropping rows with NaN
print(df_with_nan.dropna())

# Grouping
df['Department'] = ['HR', 'Tech', 'Tech', 'HR']
print(df.groupby('Department').mean(numeric_only=True))  # mean salary/age per department

# Merging DataFrames
df2 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Eve'],
    'Bonus': [20000, 150000, 5000]
})

# Left join on 'Name'
print(pd.merge(df, df2, on='Name', how='left'))

# Time series functionality

data_rng = pd.date_range(start='1/1/2023', end='1/10/2023', freq='D')  # Daily frequency

ts = pd.DataFrame(data_rng, columns=['date'])
ts['value'] = np.random.randint(0, 100, size=(len(data_rng)))  # random ints 0-99

ts.set_index('date', inplace=True)

# Resample to 3-day periods and take mean
print(ts.resample('3D').mean())
```
